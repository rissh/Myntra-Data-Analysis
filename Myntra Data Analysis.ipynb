```python
#
Pincode = []
State = []
for i in df['seller_address'][:10]:
    State.append(i.split(',')[-1].split('-')[0].strip())
    Pincode.append(i.split(' ')[-1])
```

```python
# Result
print(State, Pincode)
```

```python
#
import numpy as np
Pincode = []
State = []
for i in df['seller_address']:
    
    if(i != 'Not Mentioned'):
        State.append(i.split(',')[-1].split('-')[0].strip())
        Pincode.append(i.split(' ')[-1])
        
    else:
        Pincode.append(np.nan)
        State.append(np.nan)
```
